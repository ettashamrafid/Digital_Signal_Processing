```python
import math,random
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
```

```python
rate, data = wavfile.read('conversation.wav')
print(rate)
```

```python
duration = len(data)/rate
print(duration) 
```

```python
t = np.linspace(0,duration,len(data))
plt.plot(t,data)
plt.xlabel('Time,s')
plt.ylabel('Amplitude')
plt.title('Plotting Audio Signal in Time Domain')
```

```python
powerSpectrum, frequenciesFound, time, imageAxis = plt.specgram(data[:,0], Fs=rate)
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

```python
signal = data[:rate,0]
Signal = fft(signal)
plt.plot(np.abs(Signal[:rate//2]), lw=0.15)
plt.xlabel("frequency, Hz")
plt.ylabel("Amplitude, units")
plt.show()
```
