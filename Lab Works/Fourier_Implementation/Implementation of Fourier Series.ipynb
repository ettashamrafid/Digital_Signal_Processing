```python
import numpy as np
import scipy 
import matplotlib.pyplot as plt
```

```python
n= np.linspace(0,360,360)
curve_10= np.sin(2*np.pi*10*n)
```

```python
plt.figure(figsize=(40,10))
plt.plot(n, curve_10)
```

```python
curve_75=np.sin(2*np.pi*75*n)
```

```python
plt.figure(figsize=(40,10))
plt.plot(n, curve_75)
```

```python
curve=curve_10+curve_75
```

```python
plt.figure(figsize=(40,10))
plt.plot(n, curve)
```

```python
from scipy.fft import fft, fftfreq
N = 360

yf = fft(curve)
xf = fftfreq(N, 1 / 360)

plt.plot(abs(xf), np.abs(yf))
plt.grid()
plt.show()
```
