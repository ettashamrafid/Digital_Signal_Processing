## Task-1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_signal(freq1, freq2, duration, sampling_rate):
    t = np.arange(0, duration, 1/sampling_rate)
    signal = np.sin(2 * np.pi * freq1 * t) + np.sin(2 * np.pi * freq2 * t)
    return t, signal
```

```python
def dft(signal):
    N = len(signal)
    n = np.arange(N)
    k = n.reshape((N, 1))
    e = np.exp(-2j * np.pi * k * n / N)
    dft_result = np.dot(e, signal)
    return dft_result
```

```python
sampling_rate = 1000  # Hz
signal_duration = 1  # second
```

```python
freq1 = 100  # Hz
freq2 = 200  # Hz
t, signal = generate_signal(freq1, freq2, signal_duration, sampling_rate)
```

```python
dft_result = dft(signal)
```

```python
freq_axis = np.fft.fftfreq(len(signal)) * sampling_rate
magnitude_spectrum = np.abs(dft_result)
plt.plot(abs(freq_axis), magnitude_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum')
plt.grid(True)
plt.show()
```

## Task-2

```python
import scipy.io.wavfile as wavfile
```

```python
audio_file = '2530_AV.wav'  
Fs, y = wavfile.read(audio_file)
```

```python
t = np.arange(0, len(y)) / Fs
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Waveform of the Audio Signal')
```

```python
N = len(y)
Y = np.fft.fft(y)
magnitude_spectrum = np.abs(Y[:N//2+1])
frequencies = np.linspace(0, Fs/2, N//2+1)
```

```python
plt.subplot(2, 1, 2)
plt.plot(frequencies, magnitude_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Magnitude Spectrum of the Audio Signal')
plt.show()
```

```python
max_mag_idx = np.argmax(magnitude_spectrum)
dominant_frequency = frequencies[max_mag_idx]
print(f"Dominant frequency component: {dominant_frequency:.2f} Hz")
```
