# The acceleration X data

```python
def print_plt(file_link):
    for i in file_link:
        df_x=pd.read_csv(i, header=None, delim_whitespace=True)
        df_x['Average']=df_x.mean(axis=1)
        #plt.figure(figsize=(10,6))
        plt.plot(df_x['Average'])
```

```python
def triple_data(file_link):
    for i in file_link:
        df_x=pd.read_csv(i, header=None, delim_whitespace=True)
        data_arr=np.array(df_x)
        plt.plot(data_arr[1,])
```

```python
file_link= ['train/Inertial_Signals/body_acc_x_train.txt', 'train/Inertial_Signals/body_acc_y_train.txt', 'train/Inertial_Signals/body_acc_z_train.txt']
```

```python
print_plt(file_link)
```

```python
triple_data(file_link)
```
