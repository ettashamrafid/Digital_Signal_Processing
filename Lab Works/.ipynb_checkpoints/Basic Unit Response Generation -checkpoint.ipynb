```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import math, random
```

# Generating a Sine Curve

```python
amp=1
f=2
phi=0

t=np.linspace(0,1,100)
x= amp* np.sin(2*np.pi*f*t + phi)
```

```python
plt.plot(t,x)
plt.title('Continuous Time-Sine Wave')
plt.xlabel('Time Interval')
plt.ylabel('Amplitude')
```

# Generating a Cosine Curve

```python
y= amp* np.cos(2*np.pi*f*t + phi)
```

```python
plt.plot(t,y)
plt.title('Continuous Time-Cosine Wave')
plt.xlabel('Time Interval')
plt.ylabel('Amplitude')
```

# Generating using Random Values

```python
t_rand= np.linspace(0,1,100)
z=np.random.normal(0,1,100)
```

```python
plt.plot(t_rand, z)
plt.title('Continuous Random-Time Wave')
plt.xlabel('Time Interval')
plt.ylabel('Amplitude')
```

# Unit Step

```python
samp_range=np.linspace(-2,7,10)
curve=[]
for i in range(len(samp_range)):
    if samp_range[i]>=0:
        curve.append(1)
    else:
        curve.append(0)
```

```python
print(samp_range)
```

```python
plt.stem(samp_range, curve)
```

# Unit Ramp

```python
curve_ramp=[]
for i in range(len(samp_range)):
    if samp_range[i]<0:
        curve_ramp.append(0)
    else:
        curve_ramp.append(samp_range[i])
```

```python
plt.stem(samp_range, curve_ramp)
```

# Unit Impulse

```python
unit_impulse=[]
for i in range(len(samp_range)):
    if samp_range[i]==0:
        unit_impulse.append(1)
    else:
        unit_impulse.append(0)
```

```python
plt.stem(samp_range, unit_impulse)
```

# Exponential

```python
expo=[]
for i in range(len(samp_range)):
    expo.append(np.exp(samp_range[i]))
```

```python
plt.stem(samp_range, expo)
```

## Determining whether the function is even function or not

```python
curve_range=np.linspace(-50,50,100)
curve_range
```

```python
curve= amp* np.cos(2*np.pi*f*curve_range + phi)
length=len(curve)
print(curve)
```

# For Even Function

```python
temp=0
for i in range(int(length/2)):
    if curve[i]!=curve[length-1-i]:
        temp=1
        break
```

```python
if temp==0:
    print('Even Function')
else:
    print('Not Even function')
```

# For Odd Function

```python
temp=0
for i in range(int(length/2)):
    if curve[i]!= - curve[length-1-i]:
        temp=1
        break
```

```python
if temp==0:
    print('Odd Function')
else:
    print('Not Odd function')
```
